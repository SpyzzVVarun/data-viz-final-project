```python
import pandas as pd
import os
from tqdm import tqdm
import warnings
warnings.filterwarnings("ignore")
```

```python
file_dir = "air_quality_data/"
```

```python
stations_info = pd.read_csv(file_dir + "stations_info.csv")
stations_info.head()
```

```python
cols = set()
for file in os.listdir(file_dir):
    if file == "stations_info.csv":
        continue
    inter_data = pd.read_csv(file_dir + file)
    for col in list(inter_data.columns):
        cols.add(col)
```

```python
data = pd.DataFrame(columns = list(cols))
data
```

```python
for file in tqdm(os.listdir(file_dir)):
    if file == "stations_info.csv":
        continue
    inter_data = pd.read_csv(file_dir + file)
    inter_data['Date'] = pd.to_datetime(inter_data['From Date'])
    inter_data = inter_data.drop(["From Date", "To Date"], axis = "columns")
    inter_data = inter_data.groupby(pd.Grouper(key='Date', freq='D')).mean().reset_index()
    
    for col in list(stations_info.loc[stations_info["file_name"] == file.split(".")[0]].columns):
        if col in ["file_name", "start_month", "start_month_num", "start_year"]:
            continue
        inter_data[col] = stations_info.loc[stations_info["file_name"] == file.split(".")[0], col].item()
    data = pd.concat([data, inter_data])
```

```python
data.to_csv("raw_daily_air_quality_data.csv", index = False)
data
```
