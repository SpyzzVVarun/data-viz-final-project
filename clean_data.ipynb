```python
import pandas as pd
import os
from tqdm import tqdm
import warnings
warnings.filterwarnings("ignore")
```

```python
file_dir = "air_quality_data/"
```

```python
stations_info = pd.read_csv(file_dir + "stations_info.csv")
stations_info.head()
```

```python
clean_data = pd.DataFrame(columns=['Date', 'state',	'city', 'agency', 'station_location',  
                                   'CO (ug/m3)', 'PM10 (ug/m3)', 'PM2.5 (ug/m3)', 'NO (ug/m3)',
                                    'NO2 (ug/m3)', 'NOx (ppb)','Ozone (ug/m3)', 'SO2 (ug/m3)', 'AT (degree C)',  'RH (%)'])
```

```python
for file in tqdm(os.listdir(file_dir)):
    if file == "stations_info.csv":
        continue
    inter_data = pd.read_csv(file_dir + file)
    inter_data['Date'] = pd.to_datetime(inter_data['From Date'])
    inter_data = inter_data.drop(["From Date", "To Date"], axis = "columns")
    inter_data = inter_data.groupby(pd.Grouper(key='Date', freq='D')).mean().reset_index()
    for col in list(inter_data.columns):
        # if "CO" in col.split("(")[0]:
        #     if col == 'CO (mg/Nm3)':
        #         inter_data['CO (ug/m3)'] = 1e3*inter_data[col]
        #         inter_data = inter_data.drop(col, axis = 'columns')
        #     elif col == 'CO (mg/m3)':
        #         inter_data['CO (ug/m3)'] = 1e3*inter_data[col]
        #         inter_data = inter_data.drop(col, axis = 'columns')
        #     elif col == 'CO (ng/m3)':
        #         inter_data['CO (ug/m3)'] = 1e-3*inter_data[col]
        #         inter_data = inter_data.drop(col, axis = 'columns')
        
        # if "NO" in col.split("(")[0]:
        #     if col == 'NO (mg/m3)':
        #         inter_data['NO (ug/m3)'] = 1e3*inter_data[col]
        #         inter_data = inter_data.drop(col, axis = 'columns')
        #     elif col == 'NOx (ppm)':
        #         inter_data['NOx (ppb)'] = 1e3*inter_data[col]
        #         inter_data = inter_data.drop(col, axis = 'columns')

        if col in ['CO (mg/Nm3)',
                   'CO (mg/m3)',
                   'CO (ng/m3)',
                   'NO (mg/m3)',
                   'NOx (ppm)',
                   'AT ()',
'AT (degree)',
 'AT (ug/m3)','BP ()',
'BP (W/mt2)',
 'BP (mg/m3)',
 'BP (mmHg)',
 'BP (mmHg).1',
 'BP (mmHg).2',
 'BP (mmHg).3',
 'BP (mmHg).4',
 'BP (mmHg).5',
 'BP (mmHg).6',
 'BP (mmHg).7', 
 'CH4 ()',
 'Eth-Benzene ()',
 'Eth-Benzene (ug/m3)',
 'MP-Xylene ()',
 'MP-Xylene (ug/m3)',
  'NMHC ()',
 'NMHC (ug/m3)',
 'Benzene ()',
 'Benzene (mg/m3)',
 'Benzene (ug/m3)',
  'Xylene ()',
 'Xylene (ug/m3)',
 'HCHO (ug/m3)',
 'Hg (ug/m3)',
 'MH (m)',
 'O Xylene (ug/m3)',
  'Toluene ()',
 'Toluene (ug/m3)',
 'VWS (degree)',
 'VWS (m/s)',
 'Variance (n)',
  'WD ()',
 'WD (deg)',
 'WD (degree C)',
 'WD (degree)',
  'Gust (kl/h)',
 'Gust (km/hr)',
 'Gust (m/s)',
 'NH3 ()',
 'NH3 (ppb)',
 'NH3 (ug/m3)',
 'NO ()',
 'NO (ppb)',
 'NO (ppm)',
 'NOx (ug/m3)',
 'Ozone ()',
 'Ozone (ppb)',
 'CO2 (mg/m3)',
 'Power (W)',
 'RF ()',
 'RF (m/s)',
 'RF (mm)',
 'RF (mm).1',
 'RF (mm).2',
 'RF (mm).3',
 'RF (mm).4',
 'RF (mm).5',
 'RF (mm).6',
 'RF (mm).7',
 'RH ()',
 'RH (W/mt2)',
 'RH (degree)',
 'SO2 ()',
 'SPM (ug/m3)'
 'SR ()',
 'SR (W/mt2)',
 'SR (ug/m3)',
 'THC ()',
 'THC (ug/m3)',
 'Temp ()',
 'Temp (degree C)',
 'Temp (ug/m3)',
 'WS ()',
 'WS (m/s)',
 'WS (ug/m3)'
]:
            inter_data = inter_data.drop(col, axis = 'columns')
        
    # print(inter_data.columns)

    for col in list(stations_info.loc[stations_info["file_name"] == file.split(".")[0]].columns):
        if col in ["file_name", "start_month", "start_month_num", "start_year"]:
            continue
        inter_data[col] = stations_info.loc[stations_info["file_name"] == file.split(".")[0], col].item()

    # if (len(clean_data.columns) == len(inter_data.columns)):
    clean_data = pd.concat([clean_data, inter_data])
```

```python
clean_data.columns
```

```python
clean_data = clean_data.reset_index(drop = True)
clean_data = clean_data.drop(['SPM (ug/m3)', 'SR ()', 'CH4 (ug/m3)'], axis = "columns")
```

```python
clean_data.to_csv("clean_daily_air_quality_data.csv", index = False)
clean_data
```
